```python
import matplotlib.pyplot as plt
import numpy as np
from dataset.landsat_ts import LandsatSpectralDataset

np.set_printoptions(suppress=True, precision=4)
```

```python
DATASET_PATH = "../data/Landsat_NDVI_time_series_1984_to_2024.tif"

# Assuming dataset_path and window_size are defined
dataset = LandsatSpectralDataset(DATASET_PATH, window_size=5)

# Selecting 3 random samples
random_indices = np.random.choice(len(dataset), 3)
samples = [dataset[i] for i in random_indices]
```

```python
for i, (X, y) in enumerate(samples):
    print(f"Sample {i+1}: X (NDVI, Year): {X.cpu().numpy()}, Y (NDVI, Year): {y.cpu().numpy()}")
```

```python
# Plotting
for i, (X, y) in enumerate(samples):
    plt.figure(figsize=(10, 5))
    plt.plot(X.cpu().numpy()[:,1], X.cpu().numpy()[:,0], label='Data (X)', color='blue', linestyle='-', marker='o')
    plt.scatter(y.cpu().numpy()[1], y.cpu().numpy()[0], label='Next Value (y)', color='red')
    plt.ylim(0, 1)
    plt.xlabel('Year')
    plt.ylabel('NDVI')
    plt.title(f'Sample {i+1}')
    plt.legend()
    plt.text(0.5, 0.95, 'Winter (0)', horizontalalignment='center', transform=plt.gca().transAxes, color='green')
    plt.text(0.5, 0.9, 'Summer (0.5)', horizontalalignment='center', transform=plt.gca().transAxes, color='green')
    plt.show()
```
