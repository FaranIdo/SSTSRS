```python
import matplotlib.pyplot as plt
import numpy as np
from dataset.landsat_ts import LandsatSpectralDataset
```

```python
DATASET_PATH = "../data/Landsat_NDVI_time_series_1984_to_2024.tif"

# Assuming dataset_path and window_size are defined
dataset = LandsatSpectralDataset(DATASET_PATH, window_size=5)

# Selecting 3 random samples
random_indices = np.random.choice(len(dataset), 3)
samples = [dataset[i] for i in random_indices]
```

```python
for i, (X, y) in enumerate(samples):
    print(f"Sample {i+1}: X - {X.cpu().numpy()}, Predicted Value - {y.cpu().numpy()}")
```

```python
# Plotting
for i, (X, y) in enumerate(samples):
    plt.figure(figsize=(10, 5))
    plt.plot(range(len(X.cpu().numpy())), X.cpu().numpy(), label='Data (X)', color='blue', linestyle='-', marker='o')
    plt.scatter(5, y.cpu().numpy(), label='Next Value (y)', color='red')
    plt.ylim(0, 1)
    plt.xlabel('Time Steps')
    plt.ylabel('NDVI')
    plt.title(f'Sample {i+1}')
    plt.legend()
    plt.show()
```
